## 2.  Scrape a single page of non-tabular data



Scrape the following info from <a href="https://www.addictioncenter.com/treatment/top-10-drug-treatment-centers/">this page</a>:
    
- Name of rehab center
- Location
- Phone number
- URL

Export the data into a csv file called ```rehabs.csv```.


````python
url_2 = "https://www.addictioncenter.com/treatment/top-10-drug-treatment-centers/"
````

````python
response = requests.get(url_2)
print(response)
response.status_code
````

````python
soup_2 = BeautifulSoup(response.text,'html.parser')
````

````python
type(soup_2)
````

````python
print(soup_2.prettify)
````

````python
all_info = soup_2.find("div", class_="RehabList")
all_info
````

````python
rehabs = all_info.find_all("div",class_ = "RehabListItem__header")
rehabs
````

````python
names = [name.find('h3').get_text().strip() for name in rehabs]
names
````

````python
locations = [location.find('p').get_text().strip().replace("Locations in ","") for location in rehabs]
locations
````

````python
numbers = [number.find('a', class_ = "btn btn-primary btn-small").get_text().strip() for number in rehabs]
numbers
````

````python
links = [link.find('a', class_ = "btn btn-outline btn-small").get("href") for link in rehabs]
links
````

````python
df = pd.DataFrame(list(zip(names, locations, numbers, links)), 
               columns =['rehab_name', "location",'phone_number', 'link'])

df
````

````python
file_name = "rehabs.csv"
df.to_csv(file_name, encoding='utf-8', index=False)
````
