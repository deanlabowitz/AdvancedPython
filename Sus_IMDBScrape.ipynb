```python
## import libraries
from bs4 import BeautifulSoup
import pandas as pd
import requests 
from icecream import ic
```

```python
url = "https://www.imdb.com/chart/top?sort=rk,asc&mode=simple&page=1"
```

```python
response = requests.get(url)
print(response)
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#ic(soup)
```

```python
first_target = soup.find_all("td",class_="titleColumn")
first_target
```

```python
title = [target.find("a").get_text() for target in first_target]
title
```

```python
len(title)
```

```python
title[249] = "Tumbbad"
```

```python
title
```

```python
rank = [int(target.get_text().strip().replace(".\n","")[0:3]) for target in first_target]
rank
```

```python
year = [int(target.find("span").get_text().replace(")","").replace("(","")) for target in first_target]
year
```

```python
second_target = soup.find_all("td",class_="ratingColumn imdbRating")
second_target
```

```python
rating = [float(target.get_text().replace("\n","")) for target in second_target]
rating
```

```python
df = pd.DataFrame(list(zip(rank, title, year, rating)), 
               columns =['rank', "title",'year', 'rating'])
df
```

```python
def create_csv(df_name, file_name):
    '''
    Export your dataframe as a csv
    argument 1 = your dataframe name
    argument 2 = your file name as a string. Must include .csv
    For example "my_data.csv"
    '''
    df_name.to_csv(file_name, encoding='utf-8', index=False )
    print(f"{file_name} is in your local folder!")
```

```python
create_csv(df,"top_movies.csv")
```
