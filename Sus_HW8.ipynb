````python
from bs4 import BeautifulSoup  ## scrape info from web pages
import requests ## get web pages from server
import time # time is required. we will use its sleep function
from random import randrange # generate random numbers
from icecream import ic
````

````python
url = "https://dof.ca.gov/Forecasting/Economics/Economic_and_Revenue_Updates/"
````

````python
response = requests.get(url, verify=False)
ic(response.status_code)
soup = BeautifulSoup(response.text, 'html.parser')
#ic(soup)
````

````python
pdf_holder = soup.find_all("div", class_="quarter")
pdf_holder
````

````python
all_a = [file.find_all("a")for file in pdf_holder]
all_a
````

````python
import itertools
````

````python
big_list_holder = list(itertools.chain(*all_a))
big_list_holder
````

````python
links = [file.get("href") for file in big_list_holder]
links
````

````python
base_url = "https://dof.ca.gov/Forecasting/Economics/Economic_and_Revenue_Updates/"
link_combos_all = [base_url + all_link.get("href") for all_link in big_list_holder]
link_combos_all
````

````python
import wget
````

````python
## if you get an error in the wget step come back and run this step
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
````

````python
link_number = len(link_combos_all)
link_count = 1

for link in link_combos_all:
    ic(link)
    print(f"Downloaded link {link_count} of {link_number}")
    link_count += 1
    wget.download(link)
    snoozer = randrange(30,60) # for outside sites, use 30 - 60 seconds as a timer
    print(f"Snoozing for {snoozer} seconds before next link")
    time.sleep(snoozer)
````

````python
def download_pdfs(all_links):
    '''
    download files from a website
    argument 1 = all links to be downloaded
    '''
    link_number = len(all_links)
    link_count = 1

    for link in all_links:
        ic(link)
        print(f"Downloaded link {link_count} of {link_number}")
        link_count += 1
        wget.download(link)
        snoozer = randrange(30,60) # for outside sites, use 30 - 60 seconds as a timer
        print(f"Snoozing for {snoozer} seconds before next link")
        time.sleep(snoozer)
````

````python
download_pdfs(link_combos_all)
````
